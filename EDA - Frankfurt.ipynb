# Plotting Restaurants, Cafes and Bars across Frankfurt

```python
ffm_gdf = ox.geocoder.geocode_to_gdf('Frankfurt, DE')

# Plot axis
fig, ax = plt.subplots(figsize = (15,10))

# Visualise both on the plot
ffm_gdf.plot(ax = ax, alpha = 0.5)

restaurants.plot(ax = ax, markersize = 1, color = 'yellow', alpha = 0.8, label = 'Restaurants')
bars.plot(ax = ax, markersize = 1, color = 'red', alpha = 0.8, label = 'Bars')
cafes.plot(ax = ax, markersize = 1, color = 'black', alpha = 0.8, label = 'Cafes')


plt.legend()
```

# Merged Plot of Street and Amenities

```python
ffm_gdf = ox.geocode_to_gdf('Frankfurt, DE')

graph = ox.graph_from_place('Frankfurt, Germany', network_type='drive')

fig, ax = plt.subplots(figsize=(15, 10))

ox.plot_graph(graph, node_color='w', node_edgecolor='k', node_size=0, edge_color='gray', ax=ax, show=False)

ffm_gdf.plot(ax=ax, alpha=0.5)
restaurants.plot(ax=ax, markersize=10, color='yellow', alpha=0.8, label='Restaurants')
bars.plot(ax=ax, markersize=10, color='red', alpha=0.8, label='Bars')
cafes.plot(ax=ax, markersize=10, color='black', alpha=0.8, label='Cafes')

plt.legend()

plt.show()
```

# Public Transport - Tram station across Frankfurt

```python
# Define the city or location of interest
place_name = "Frankfurt, Germany"  # Replace with your desired location

# Retrieve the public transport network using osmnx
public_transport_graph = ox.graph_from_place(place_name,retain_all=False, truncate_by_edge=True, simplify=True,
                        custom_filter='["railway"~"tram|rail"]')


```

```python
# Set up a plot axis
fig, ax = plt.subplots(figsize=(15, 10))

# Plot the public transport network
ox.plot_graph(public_transport_graph, node_color='red', node_size=100,edge_color='white',  ax=ax, show=False)


# Display the plot
plt.show()
```

# Plotting Tram Stations and Amenities

```python
ffm_gdf = ox.geocode_to_gdf('Frankfurt, DE')

graph = ox.graph_from_place('Frankfurt, Germany', network_type='drive')

fig, ax = plt.subplots(figsize=(15, 10))

ox.plot_graph(graph, node_color='w', node_edgecolor='k', node_size=0, edge_color='gray', ax=ax, show=False)
ox.plot_graph(public_transport_graph, node_color='red', node_size=100,edge_color='white',  ax=ax, show=False)

ffm_gdf.plot(ax=ax, alpha=0.5)
restaurants.plot(ax=ax, markersize=10, color='yellow', alpha=0.8, label='Restaurants')
bars.plot(ax=ax, markersize=10, color='red', alpha=0.8, label='Bars')
cafes.plot(ax=ax, markersize=10, color='black', alpha=0.8, label='Cafes')

plt.legend()

plt.show()
```

# Parking Spots across Frankfurt

```python
# Define the city or location of interest
place_name = "Frankfurt, Germany"  # Replace with your desired location

# Retrieve parking facilities using osmnx
parking_points = ox.geometries_from_place('Frankfurt, DE', tags = {'amenity':'parking'})

# Set up a plot axis
fig, ax = plt.subplots(figsize=(15, 10))

# Plot the parking facilities
parking_points.plot(ax=ax, color='blue', alpha=0.7, markersize=10)

# Display the plot
plt.title(f"Parking Facilities in {place_name}")
plt.show()
```

# Plotting Amentities, Parking spots and Tram Stations

```python
ffm_gdf = ox.geocode_to_gdf('Frankfurt, DE')

graph = ox.graph_from_place('Frankfurt, Germany', network_type='drive')

fig, ax = plt.subplots(figsize=(15, 10))

ox.plot_graph(graph, node_color='w', node_edgecolor='k', node_size=0, edge_color='gray', ax=ax, show=False)
ox.plot_graph(public_transport_graph, node_color='red', node_size=100,edge_color='white',  ax=ax, show=False)

ffm_gdf.plot(ax=ax, alpha=0.5)
restaurants.plot(ax=ax, markersize=10, color='yellow', alpha=0.8, label='Restaurants')
bars.plot(ax=ax, markersize=10, color='red', alpha=0.8, label='Bars')
cafes.plot(ax=ax, markersize=10, color='black', alpha=0.8, label='Cafes')
parking_points.plot(ax=ax, color='blue', alpha=0.7, markersize=10, label='Parking Spots')

plt.legend()

plt.show()
```
